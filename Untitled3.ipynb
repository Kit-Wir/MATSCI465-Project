```python
import numpy as np
import matplotlib.pyplot as plt

def make_disk(shape, cy, cx, r, amp=1.0):
    ky, kx = shape
    Y, X = np.ogrid[:ky, :kx]
    R = np.sqrt((Y-cy)**2 + (X-cx)**2)
    return amp * (R <= r).astype(np.float32)

def make_gaussian_peak(shape, cy, cx, sigma, amp=1.0):
    ky, kx = shape
    Y, X = np.ogrid[:ky, :kx]
    return amp * np.exp(-((Y-cy)**2 + (X-cx)**2)/(2*sigma**2)).astype(np.float32)

def rotate_points(points, angle_rad):
    R = np.array([[np.cos(angle_rad), -np.sin(angle_rad)],
                  [np.sin(angle_rad),  np.cos(angle_rad)]])
    return points @ R.T

def synthetic_4dstem(ny=128, nx=128, ky=128, kx=128, seed=0,
                     n_precip=12, precip_radius=7,
                     base_spot_amp=80, precip_spot_amp=35,
                     noise_sigma=1.5):
    """
    Creates a toy 4D-STEM dataset:
    - matrix has a set of Bragg peaks (spots)
    - precipitate regions add extra 'satellite' peaks (second-phase signature proxy)
    - orientation varies slowly across scan by rotating peak positions
    Returns:
      data4d: (ny,nx,ky,kx)
      mask_precip: (ny,nx) bool
    """
    rng = np.random.default_rng(seed)
    data4d = np.zeros((ny,nx,ky,kx), dtype=np.float32)

    # --- ground truth precipitate mask: random circles ---
    mask = np.zeros((ny,nx), dtype=bool)
    centers = rng.integers(low=15, high=min(ny,nx)-15, size=(n_precip, 2))
    Yg, Xg = np.ogrid[:ny, :nx]
    for (cy0, cx0) in centers:
        mask |= ((Yg-cy0)**2 + (Xg-cx0)**2 <= precip_radius**2)

    # --- reciprocal space center ---
    dcy, dcx = ky//2, kx//2

    # Base Bragg peak positions in reciprocal pixels (relative coords)
    base_pts = np.array([
        [0, 18], [0, -18], [18, 0], [-18, 0],
        [13, 13], [13, -13], [-13, 13], [-13, -13]
    ], dtype=np.float32)

    # Precipitate "extra peaks" (satellites)  proxy for second-phase scattering
    sat_pts = np.array([
        [0, 26], [0, -26], [26, 0], [-26, 0],
        [18, 18], [18, -18], [-18, 18], [-18, -18]
    ], dtype=np.float32)

    # Precompute a central beam disk
    central = make_disk((ky,kx), dcy, dcx, r=4, amp=200)

    # Slow orientation field across scan
    # angle varies smoothly from -theta to +theta
    theta_max = np.deg2rad(10)
    yy = np.linspace(-1, 1, ny)[:, None]
    xx = np.linspace(-1, 1, nx)[None, :]
    angle_field = theta_max * (0.6*xx + 0.4*yy)  # smooth gradient

    # Thickness-like multiplicative field (optional)
    thick_field = 0.85 + 0.3 * (xx + 1)/2  # ~0.85..1.15

    for i in range(ny):
        for j in range(nx):
            ang = angle_field[i, j]
            t = thick_field[0, j]

            dp = np.zeros((ky,kx), dtype=np.float32)
            dp += central

            # rotate base peaks
            pts = rotate_points(base_pts, ang)
            for py, px in pts:
                dp += make_gaussian_peak((ky,kx), dcy+py, dcx+px, sigma=1.6, amp=base_spot_amp)

            # precipitate adds satellites + slight diffuse ring
            if mask[i, j]:
                pts2 = rotate_points(sat_pts, ang)
                for py, px in pts2:
                    dp += make_gaussian_peak((ky,kx), dcy+py, dcx+px, sigma=2.0, amp=precip_spot_amp)
                # diffuse scattering proxy
                dp += make_gaussian_peak((ky,kx), dcy, dcx, sigma=14, amp=12)

            # apply thickness scaling + noise
            dp *= t
            dp += rng.normal(0, noise_sigma, size=(ky,kx)).astype(np.float32)
            dp = np.clip(dp, 0, None)

            data4d[i, j] = dp

    return data4d, mask

# --- Generate dataset ---
data4d, mask_gt = synthetic_4dstem(ny=128, nx=128, ky=128, kx=128, seed=0)

print("data4d:", data4d.shape, "mask:", mask_gt.shape, "gt precip fraction:", mask_gt.mean())

# visualize GT mask
plt.figure(figsize=(6,4))
plt.imshow(mask_gt, cmap="gray")
plt.title("Ground truth precipitate mask (synthetic)")
plt.axis("off")
plt.show()

# visualize mean DP
mean_dp = data4d.mean(axis=(0,1))
plt.figure(figsize=(5,5))
plt.imshow(mean_dp, origin="lower")
plt.title("Mean diffraction pattern (synthetic)")
plt.colorbar()
plt.show()

np.save("sim_data4d.npy", data4d)
np.save("sim_mask_gt.npy", mask_gt.astype(np.uint8))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

Xfeat = np.load("sim_radial_features.npy")
ygt = np.load("sim_mask_gt.npy").reshape(-1).astype(int)
ny, nx = 128, 128

Xs = StandardScaler().fit_transform(Xfeat)
Z = PCA(n_components=10, random_state=0).fit_transform(Xs)

# try k=2..5 and pick best by silhouette OR just set k=2 for simplicity
k = 2
labels = KMeans(n_clusters=k, n_init=20, random_state=0).fit_predict(Z[:, :5])

# choose which cluster = precip (higher GT fraction)
precip_frac = []
for c in range(k):
    precip_frac.append(ygt[labels==c].mean())
precip_frac = np.array(precip_frac)
precip_cluster = precip_frac.argmax()
print("Precipitate fraction per cluster:", precip_frac, "=> precip_cluster:", precip_cluster)

ypred = (labels == precip_cluster).astype(int)

# visualize
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.imshow(ygt.reshape(ny,nx), cmap="gray")
plt.title("GT precipitate mask")
plt.axis("off")
plt.subplot(1,2,2)
plt.imshow(ypred.reshape(ny,nx), cmap="gray")
plt.title("Predicted precipitate mask (unsupervised)")
plt.axis("off")
plt.tight_layout()
plt.show()

np.save("sim_labels.npy", labels)
np.save("sim_pred_mask.npy", ypred.reshape(ny,nx))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import adjusted_rand_score
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

data4d = np.load("sim_data4d.npy").astype(np.float32)
ny, nx, ky, kx = data4d.shape

cy, cx = ky//2, kx//2
Y, X = np.ogrid[:ky, :kx]
R = np.sqrt((Y-cy)**2 + (X-cx)**2)
nbins = 20
edges = np.linspace(0, R.max(), nbins+1)
masks = [(R>=edges[i]) & (R<edges[i+1]) for i in range(nbins)]

def build_features(d4):
    rad = np.stack([d4[..., m].sum(axis=-1) for m in masks], axis=-1)
    total = d4.sum(axis=(-2,-1)) + 1e-8
    rad_norm = rad / total[..., None]
    return rad_norm.reshape(ny*nx, nbins)

def cluster(Xfeat, k=2):
    Xs = StandardScaler().fit_transform(Xfeat)
    Z = PCA(n_components=10, random_state=0).fit_transform(Xs)
    return KMeans(n_clusters=k, n_init=20, random_state=0).fit_predict(Z[:, :5])

def add_poisson(d4, scale, seed=0):
    rng = np.random.default_rng(seed)
    x = np.clip(d4*scale, 0, None)
    return rng.poisson(x).astype(np.float32)

X0 = build_features(data4d)
lab0 = cluster(X0, k=2)

scales = [1.0, 0.5, 0.2, 0.1]
aris = []
for sc in scales:
    dn = add_poisson(data4d, sc, seed=0)
    Xn = build_features(dn)
    labn = cluster(Xn, k=2)
    aris.append(adjusted_rand_score(lab0, labn))

print("Dose scales:", scales)
print("ARI vs baseline:", aris)

plt.figure()
plt.plot(scales, aris, marker="o")
plt.xlabel("Dose scale (Poisson)")
plt.ylabel("ARI stability vs baseline")
plt.title("Synthetic robustness to dose reduction")
plt.grid(True)
plt.show()
```
